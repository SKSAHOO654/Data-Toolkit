2 .Write a Pandas script to find the mean of a specific column in a DataFrame ?

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Salary": [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

mean_age = df["Age"].mean()

print("Mean Age:", mean_age)
```

4 .How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")

df = df.select_dtypes(include=["number"])

corr_matrix = df.corr()
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")

plt.show()
```

8 .Create a line plot with multiple lines using Matplotlib ?

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]  # Line 1
y2 = [5, 10, 15, 20, 25]   # Line 2

plt.plot(x, y1, label="Line 1", color="blue", marker="o", linestyle="-")
plt.plot(x, y2, label="Line 2", color="red", marker="s", linestyle="--")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot Example")

plt.legend()

plt.show()
```

9 .Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold ?

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Salary": [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

filtered_df = df[df["Salary"] > 60000]

print(filtered_df)
```

10 .Create a histogram using Seaborn to visualize a distribution ?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=50, scale=15, size=200)

sns.histplot(data, bins=20, kde=True, color="blue")

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Data Distribution")

plt.show()
```

13 .Create a 3D scatter plot using Plotly ?

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    "X": np.random.randint(1, 100, 50),
    "Y": np.random.randint(1, 100, 50),
    "Z": np.random.randint(1, 100, 50),
    "Category": np.random.choice(["A", "B", "C"], 50)
})

fig = px.scatter_3d(df, x="X", y="Y", z="Z", color="Category", title="3D Scatter Plot")

fig.show()
```
