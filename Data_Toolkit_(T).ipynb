1 .What is NumPy, and why is it widely used in Python ?

NumPy (Numerical Python) is a powerful open-source library in Python used for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these data structures efficiently.



- Why is NumPy Widely Used?
 - Performance: NumPy is implemented in C and optimized for speed, making it much faster than standard Python lists for numerical operations.
 - Multi-Dimensional Arrays: Provides an efficient ndarray (N-dimensional array) object that allows easy manipulation of data.
 - Mathematical Functions: Supports a wide range of mathematical operations, including linear algebra, statistics, and Fourier transforms.
 - Broadcasting: Allows operations on arrays of different shapes without explicit looping, making code more efficient.
 - Memory Efficiency: Uses contiguous memory blocks, reducing overhead and improving cache performance.
 - Interoperability: Works well with other libraries like Pandas, SciPy, Matplotlib, and machine learning frameworks like TensorFlow and PyTorch.
 - Vectorization: Eliminates the need for explicit loops in Python by applying operations element-wise, leading to significant performance improvements.

2 .How does broadcasting work in NumPy ?

- Compare shapes from right to left: If dimensions match or one of them is 1, they are compatible.
- Expand dimensions where needed: If one array has a lower-dimensional axis (size 1), NumPy stretches it to match the other array.
- Apply element-wise operations: After broadcasting, NumPy performs the operation as if the arrays were of the same shape.

3 .What is a Pandas DataFrame ?

17 .Explain the difference between apply() and map() in Pandas ?

1. map()  Works Only on Series (Element-Wise)
- Used for applying functions to a single Pandas Series.
- Works element-wise, meaning it applies a function to each value individually.
- Can take built-in functions, custom functions, or dictionaries for mapping.

2. apply()  Works on Both DataFrames & Series
- Used for applying functions to either rows or columns of a DataFrame.
- Works on Series (like map()) or entire DataFrames.
- Can apply row-wise (axis=1) or column-wise (axis=0).

21 .Why is NumPys array slicing faster than Pythons list slicing ?

1. NumPy Uses a Continuous Memory Block
- NumPy arrays are stored in a contiguous block of memory (like C arrays), whereas Python lists store pointers to separate objects in memory.

 - Result: NumPy can access and slice elements directly without dereferencing multiple pointers.

2. NumPy Uses Views Instead of Copies
- Slicing a NumPy array creates a view (reference to the same data), whereas slicing a Python list creates a new list (copy of data).

 - Result: No extra memory usage in NumPy slicing!

3. NumPy is Optimized with C and Vectorization
- NumPy is implemented in C and Fortran, allowing direct memory access and vectorized operations without Python loops.

 - Result: Faster execution since Python lists require an interpreter overhead for element-wise operations.

4. NumPy Leverages SIMD and Parallel Computing
- NumPy automatically optimizes computations using Single Instruction Multiple Data (SIMD) and parallel processing.
- Python lists, on the other hand, rely on Python loops, which are much slower.
